```
jdbc_url = "jdbc:sqlserver://capstoneprojectnewserver.database.windows.net:1433;database=capstonesqldatabase;encrypt=true;trustServerCertificate=false;"

# Define the connection properties
connection_properties = {
    "user": "sqladmin",
    "password": "PASSword@1234",
    "driver": "com.microsoft.sqlserver.jdbc.SQLServerDriver"
}

query = "(SELECT * FROM DataFromDatabricksNotebook) AS my_table"
airbnb_df = spark.read.jdbc(url=jdbc_url, table=query, properties=connection_properties)
```

```
airbnb_df.createOrReplaceTempView("airbnb")
```

```
%sql
SELECT COUNT(DISTINCT listing_id) as TotalListingIds
FROM airbnb
```

```
%sql
SELECT COUNT(DISTINCT city) as TotalCities
FROM airbnb
```

```
%sql
SELECT COUNT(DISTINCT review_id) FROM airbnb
```

```
%sql
SELECT city, avg(review_scores_rating) AS AverageRating
FROM airbnb
WHERE city IS NOT NULL
GROUP BY city
HAVING AverageRating<=100
ORDER BY AverageRating DESC
```

```
%sql
SELECT city, AVG(price) AS average_price, MIN(price) AS min_price, MAX(price) AS max_price
FROM airbnb
GROUP BY city
ORDER BY average_price DESC
```

```
%sql 
SELECT city, COUNT(listing_id) AS total_listings
FROM airbnb
WHERE city IS NOT NULL
GROUP BY city
ORDER BY total_listings DESC
```

```
%sql
SELECT city, SUM(CASE WHEN instant_bookable = 'True' THEN 1 ELSE 0 END) AS instant_bookable_count
FROM airbnb
WHERE city IS NOT NULL
GROUP BY city
HAVING instant_bookable_count>0
```

```
%sql
SELECT property_type, COUNT(listing_id) AS count
FROM airbnb
GROUP BY property_type
ORDER BY count DESC
LIMIT 20
```

##### Average price by city (Bar graph)

##### Distribution of Review Scores for Cleanliness (Histogram)

```
%sql
SELECT city, avg(review_scores_cleanliness) AS AverageCleanRate
FROM airbnb
WHERE review_scores_cleanliness IS NOT NULL AND review_scores_cleanliness<=15 AND city IS NOT NULL
GROUP BY city
ORDER BY AverageCleanRate;
```

##### Number of Listings per Property Type (Pie Chart)

##### Average Review Scores by Room Type (Multi-Series Bar Chart)

##### Listings with Host Profile Picture and Identity Verified (Stacked Bar Chart)

```
%sql
SELECT host_has_profile_pic, host_identity_verified, COUNT(*) AS number_of_listings
FROM airbnb
GROUP BY host_has_profile_pic, host_identity_verified;
```

##### Correlation Between Price and Review Scores (Scatter Plot)

```
%sql
SELECT price, review_scores_rating
FROM airbnb
WHERE price < 1000 AND review_scores_rating<100 AND review_scores_rating>50 ;
```

```
%sql
SELECT review_scores_cleanliness, review_scores_rating
FROM airbnb
WHERE review_scores_rating<100 AND review_scores_rating IS NOT NULL AND review_scores_cleanliness<20;
```

```
%sql
SELECT review_scores_communication, review_scores_rating
FROM airbnb
WHERE review_scores_rating<100 ;
```

```
%sql
SELECT element AS value, COUNT(*) AS count
FROM airbnb LATERAL VIEW EXPLODE(split(amenities, ',')) t AS element
GROUP BY element
ORDER BY count DESC
LIMIT 1000;
```
